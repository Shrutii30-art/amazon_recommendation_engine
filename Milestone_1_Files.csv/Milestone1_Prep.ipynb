```python
import pandas as pd

# 1. LOAD DATA
# This uses the new electronics dataset you renamed
df = pd.read_csv('amazon_product.csv')

# 2. DATA EXPLORATION (Milestone 1 Task)
# Let's see how many products and what columns we have
print(f"Total products in new dataset: {len(df)}")
print(f"Columns found: {df.columns.tolist()}")

# 3. PRE-PROCESSING & CLEANING
# We need to create a clean productId. Using the row index is the safest way
# to ensure the matrix and images match perfectly later.
df['productId'] = df.index.astype(str)

# Keep the essential columns for your placement project UI
# 'image' is the column containing the photo links you need.
essential_cols = ['productId', 'name', 'main_category', 'image', 'ratings', 'actual_price']
df_clean = df[essential_cols].copy()

# Remove any products that are missing names or images to keep the catalog professional
df_clean.dropna(subset=['name', 'image'], inplace=True)

# 4. SAVE CLEANED DATA
# Overwriting the file with the clean version for Milestone 2
df_clean.to_csv('amazon_products.csv', index=False)

print("\n✅ Milestone 1 Complete!")
print(f"Cleaned dataset saved. Ready with {df_clean.shape[0]} products.")
print("The 'productId' now ranges from 0 to", len(df_clean)-1)
```
